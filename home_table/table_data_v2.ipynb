```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
DATA_PATH = '/Users/dafneculha/Desktop/project-2023-choo-choo-data-darlings/home_table/transports_v2.parquet'
transport_data = pd.read_parquet(DATA_PATH)
```

```python
transport_df = pd.DataFrame(transport_data)
```

```python
transport_df
```

```python
transport_df = transport_df.drop_duplicates()
## nothing changes so it's good
```

```python
transport_df
```

```python
agg_dict = {'mean_arrival_delay': 'mean',
            'mean_departure_delay': 'mean',
            'median_arrival_delay': 'median',
            'median_departure_delay': 'median',
            'std_arrival_delay': 'std',
            'std_departure_delay': 'std',
            'n_arrival_delay': 'sum',
            'n_departure_delay': 'sum',
            'n_cancelled': 'sum',
            'n_through_trip': 'sum',
            'n_additional_trip': 'sum',
            'n_entries': 'sum'}
```

```python
grouped_df = transport_df.groupby('stop_id').agg(agg_dict)
```

```python
grouped_df
```

```python
grouped_df.head()
```

```python
# Load the other df
df_names = pd.read_csv('/Users/dafneculha/Desktop/project-2023-choo-choo-data-darlings/data/processed/stops.csv', usecols=['stop_id', 'stop_name', 'city', 'canton'], index_col='stop_id')

```

```python
df_names 
```

```python


merged_df = df_names.merge(grouped_df, on='stop_id')
```

```python
merged_df.head()
```

```python
df_tosave = merged_df.reset_index().to_json(orient='records')

#with open('table_df.json', 'w') as f:
#    f.write(df_tosave)
```

```python

hist_values, bins, _ = plt.hist(merged_df['mean_arrival_delay'], bins=100)  # Adjust the number of bins as needed

plt.xlabel('Mean Arrival Delay')
plt.ylabel('Frequency')
plt.title('Histogram of Mean Arrival Delay')

histogram_settings = {
    'num_bins': len(bins) - 1,  # Number of bins
    'bin_size': (bins[1] - bins[0]).tolist(),  # Bin size
    'bins': bins.tolist(),  # Convert the bins array to a list
    'hist_values': hist_values.tolist()  # Convert the hist_values array to a list
}

with open('mean_arrival_delay.json', 'w') as f:
    json.dump(histogram_settings, f)

plt.show()
```
