```python
import pandas as pd
```

```python
# read yaaas.csv
df = pd.read_csv("yaaas.csv")

# read network_nodes.csv
df_nodes = pd.read_csv("network_nodes.csv")

# read network_edges.csv
df_edges = pd.read_csv("network_edges.csv")
```

```python
# reaname columns by removing spaces and capital letters
df.columns = df.columns.str.replace(" ", "_").str.lower()
# drop label
df = df.drop(columns=['label'])
```

```python
# drop indegree outdegree degree from df_nodes
df_nodes = df_nodes.drop(columns=["indegree", "outdegree", "degree"])
```
